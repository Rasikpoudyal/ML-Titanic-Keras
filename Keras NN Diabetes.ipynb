```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from sklearn.cluster import KMeans
from sklearn import datasets
import pandas as pd

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
dataset = np.loadtxt('C:\\Users\Rasik\Documents\pima-indians-diabetes\pima-indians-diabetes.csv', delimiter = ",")
X = dataset[:,0:8,]
Y = dataset[:,8,]
```

```python
print(X.shape)
```

```python
print(Y.shape)
```

```python
#Sequential model, Dense = dense neurons, units = numners, input_dim = number of data? 
#activation='relu' = Rectifier f(x) =log(1+exp x),
#where x in the input to a neuron


model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu' ))
model.add(Dense(15, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(2, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

```

```python
model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ['accuracy'])
```
